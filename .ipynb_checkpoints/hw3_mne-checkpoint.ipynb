```python
from category_encoders import TargetEncoder
```

```python
from sklearn.linear_model import Lasso, LassoCV,Ridge 
```

Делим данные на тестовые и на которых обучаться будем.

```python
_train,_test=train_test_split(data,train_size=0.8,random_state=42)
```

```python
_train.head()
```

```python
x_train=_train.drop(columns='popularity')
y_train=_train['popularity']
x_test=_test.drop(columns='popularity')
y_test=_test['popularity']
```

```python
categ_cols=['artist','song','genre']
encoder=TargetEncoder(cols=categ_cols)
```

```python
x_train_encoded=encoder.fit_transform(x_train,y_train)
x_test_encoded=encoder.transform(x_test)
```

Проверим результат с ласо.

```python
lasso1 = Lasso(alpha=0.01,random_state=42) 
lasso1.fit(x_train_encoded,y_train) 
y_pred_train_las1=lasso1.predict(x_train_encoded)
mean_train_las1=np.mean((y_pred_train_las1-y_train)**2)
y_pred_test_las1=lasso1.predict(x_test_encoded)
mean_test_la1s=np.mean((y_pred_test_las1-y_test)**2)
```

```python
alphas = np.logspace(-6, 2, 200)
lasso_cv = LassoCV(n_alphas=len(alphas),alphas=alphas,random_state=42,cv=5) 
lasso_cv.fit(x_train_encoded,y_train) 
```

```python
y_pred_train_las_cv=lasso_cv.predict(x_train_encoded)
mean_train_las_cv=np.mean((y_pred_train_las_cv-y_train)**2)
y_pred_test_las_cv=lasso_cv.predict(x_test_encoded)
mean_test_las_cv=np.mean((y_pred_test_las_cv-y_test)**2)
```

```python
print("Mean squared error (train): %.3f" %mean_train_las_cv) 
print("Mean squared error (test): %.3f" %mean_test_las_cv) 
```

Проверим результат с Ridge.

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
ridge_model = Ridge(alpha=1.0)  
cv_scores = cross_val_score(ridge_model, x_train_encoded, y_train, cv=kf, scoring='neg_root_mean_squared_error')
print("Cross-validated RMSE:", -cv_scores.mean())
ridge_model.fit(x_train_encoded, y_train)
pred = ridge_model.predict(x_test_encoded)
```

```python
ridge_rmse = mean_squared_error(y_test, pred, squared=False)
print("rmse score equals to:", ridge_rmse)
```
